EMPLOYEE FUNCTION

```python

# Import necessary libraries
import os
import zipfile

# Function to retrieve employee from CSV File
def get_employee_details(employee_name, df):

    try:
        # Filter the DataFrame to rows where 'EmployeeName' matches the input name
        filtered_df = df[df['EmployeeName'] == employee_name]
        if filtered_df.empty:
            return None
        employee_details = filtered_df.iloc[0].to_dict()
        return employee_details
    except KeyError as e:
        print(f"Error: DataFrame does not contain 'EmployeeName' column. {e}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Function to export employee details to a CSV file
def export_employee_details(employee_details, output_folder):

    os.makedirs(output_folder, exist_ok=True)

    df = pd.DataFrame([employee_details])

    # Define the output file path
    output_file = os.path.join(output_folder, f"{employee_details['EmployeeName']}_details.csv")

    # Export the DataFrame to CSV
    df.to_csv(output_file, index=False)
    print(f"Employee details exported to: {output_file}")

# Main function
def main():
    try:
        # Load DataFrame
        df = pd.read_csv("Total.csv")  # Load your CSV file here
    except FileNotFoundError:
        print("Error: File 'Total.csv' not found.")
        return

    # Get employee name from user input
    employee_name = input("Enter employee name: ").strip()

    # Get employee details
    employee_details = get_employee_details(employee_name, df)

    # Display results
    if employee_details:
        print(f"\nEmployee details for '{employee_name}':")
        for key, value in employee_details.items():
            print(f"{key}: {value}")

        # Export employee details to CSV
        output_folder = "Employee Profile"
        export_employee_details(employee_details, output_folder)

        # Zip the output folder
        zip_filename = "Employee_Profile.zip"
        with zipfile.ZipFile(zip_filename, 'w') as zipf:
            for root, _, files in os.walk(output_folder):
                for file in files:
                    zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), output_folder))

        print(f"\nEmployee profile saved to: {zip_filename}")
    else:
        print(f"\nEmployee '{employee_name}' not found.")

# Execute the main function if the script is run directly
if __name__ == "__main__":
    main()
```
